Text similarity detection

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

text1 = "I am going to study"
text2 = "I am going to college"

embeddings = model.encode([text1, text2], convert_to_tensor=True)
similarity_score = util.cos_sim(embeddings[0], embeddings[1])
print("Similarity score:", similarity_score.item())
```

Sentiment Analysis

```python
from transformers import pipeline

sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

text = "I am going to college"

result = sentiment_pipeline(text)
print(result)
```

Fake news classifier

```python
from transformers import pipeline

fake_news_pipeline = pipeline(
    "text-classification",
    model="mrm8488/bert-tiny-finetuned-fake-news-detection"
)

news = "The government has elected Donald Trump as the most peaceful person for this year"

result = fake_news_pipeline(news)
print(result)
```
